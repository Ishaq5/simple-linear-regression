### 1- Correlation Coefficient

But do we determine how strong is the correlation numerically?

We can use `correlation coefficient` to determine:
- if correlation is perfect, coefficient = 1 
- if correlation is strong, coefficient > 0.7
- if correlation is moderate, 0.3 < coefficient < 0.5
- if correlation is weak , coefficient < 0.3
- if no correlation , coefficient = 0

Can we have negative correlation?

Since we have positive correlation, negative correlation meaning an increase in independant variable results in decrease in dependant variable. For negatively correlated relationships, coefficient with be negative. 

### 2- Calculating correlation coefficient

The equation to calculate correlation coefficient r:


- Σ is Sigma, the symbol for "sum up"
- (xi - xbar) is each x-value minus the mean of x (called `city-L/100km`)
- (yi - ybar) is each y-value minus the mean of y (called `price`)

let us use the equation to calculate the coefficient for `price` and `city-L/100km`

```python
# calculating correlation coefficient

# mean of x and y 
x_bar = df['city-L/100km'].mean()
y_bar = df['price'].mean()

# x minus xbar 
x_xbar= df['city-L/100km']-x_bar
y_ybar = df['price']-y_bar

#sum of x_xbar multiplied y_ybar
numerator = (x_xbar*y_ybar).sum()

# squaring 'x_xbar' and 'y_ybar'
sqrt_x_xbar = x_xbar **2
sqrt_y_ybar = y_ybar **2

# calculating denomirator 
denomirator = (sqrt_x_xbar.sum() * sqrt_y_ybar.sum())**0.5

#calculating the coefficient 
coef = numerator/denomirator
print(coef)
```

## B-Simple Linear Regression

### 1- equation of the line

**Now that we can identify attributes that have strong correlation with our target `price`, let us discuss what linear regression is and how we can make a model from attributes that have strong correlation with `price`.**

We will start this tutorial considering only one attribute to predict `price`.  **`simple linear regression`.**

If you remember from school, we can describe a line in 2D using the equation of a line. 
that had the following:

- y= output value
- x= input value
- m= slope od a line
- b= y-axis intercept

Since m is the slop of the line, It describes the rate of change of y over x.

**m = change of y / change of x**

Meaning it describes for each change in x, how much y changes. 



Let us plot a line using the equation above

```python
x = np.array(list(range(1,9)))

''' let m=2 meaning for every change in x
    y will change twice '''

m = 2 
b = 1 

y = m*x + b

print(x)
print(y)
plt.plot(x,y)
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

**This equation of the line we can use to predict or know any value along this line** 

For example let predict of estimate what is the value of y if x is let's say 12?

using the equation that we have:

y = 2*(12)+1 = 25

**Ok, we uderstand the equation of the line, How is that knowledge will help us in our problem?**




We have the result already `data point of x and y` in dataset our task is to figure out `the line` that describes the relationship between car attributes and `price`. That my friend is `machine learning`. We need to find the line prameters `m` and `b` the slope and the y intercept to end up with an equation we can use to calculate future value of an attribute. 

For this example, we will use `engine-size` as x or independant attribute and `price` as y or dependent attribute. 

In the plot above, we had points (x) and the parameters m and b and needed to calculate (y).
In our dataset, we have (x) and (y) but we lack m and b. Once we obtain m and b, we will have an equation that can estimate any new point y given a point x. Our equation will be 

**`price` = m * `engine-size` + b**

### 2- Cost function 

As we discussed above, we are trying to find a line (equation) that goes through data point by finding `m` and `b` parameters of the equation. As in school, algebra can help us doing that. For example, we can take two values as points from `price` and the same for `engine-size`.

**First approach:**
- calculate `m` by (change of price) / (change of engine-size) 
- calculate `b` by price - (m * engine=sinze)

**Second approch:**

We might also just to plot a random line and somehow hope to get it in the best place be moving it 
meaning obtain an equation line then change `m` and `b` such that it moves to best spot among data points. 

**Third approch:**

Let's compine first and second approch using what is called `Cost Function`.

```python
# example of a line badly fitting data points

y_mean = [df['price'].mean() for i in range(df['price'].shape[0])]

plt.plot(df['engine-size'],y_mean,color='green')
plt.scatter(df['engine-size'],df['price'])

plt.xlabel('engine-size')
plt.ylabel('price')
plt.title('Price and Engine-Size correlation')
plt.show()
```

```python
# example of a line fitting the data
sns.regplot(df['engine-size'],df['price'])
```

In the first plot, we used the average of `price` as a line that doesn't fit the data well. For the second plot however, the line seems to be fitting data points better than the average line. To find such a line we will introduce a cost function that is known as `sums of squared errors` or `sums of squared residuals` SSE or SSR which all refere to the total difference between actual target value `price` and predict value `price hat` we will call it. 

- actual value y (points that are scattered in the plot above) 
- predicted value y^ or y hat ( the projected valued of scattered points on the line)




An error is simply the difference between the actual target value `y` and predicted value `y hat`.

For i which is any value from the column price:

**ERROR = (y)i - (y_hat)i**
**Total ERROR = sum(ERROR)**

Remember; our objective is to find and equation of a line that fits points the best. We can use the error equation to calculate all the errors for all points in price column and find the sum of all errors then based on the total errors, we choose the line with least total error. 

But we have a problems with ERROR equation above:

- As we can see from the plot above, some points are above the line while others are below the line. That will introduce an issue while calculating total errors as some errors will cancel each other. Some errors will be:
    - positive (actual value yi will be above the line)
    - negative (actual value yi will be below the line)

We can enhance the model by finding the absolute value of ERROR, forcing ERROR to be always positive value

**ABSOLUTE ERROR = | (y)i - (y_hat)i |**

Since we want to minimize `cost function` in our case ABSOLUTE ERROR, we need to find the first derivative of the error function that we have `ABSOLUTE ERROR` where the function is equal to zero. The probelm here if you remember anything from deravitives, we can not find the deravitive for undifferential equation `absolute function`. 



It pivots at x=0 and so this function is differentiable everywhere except at x=0. And so `ABSOLUTE ERROR` is not very useful in this case. 

`sums of squared errors` is exactly what we need for this problem:
- It prevents errors canceling each other
- It is differentiable everywhere along x axis



### 3-  Finding y-intercept and slope of the line


### Y_intercept `b`


**From above equation, let's minimize `sum of squared errors` SSE cost function with respect to `m` and `b` "aka" partially differentiate SSE cost function with respect to slope and y_intercept. after we obtain two derivative, we solve them for `m` the slope and `b` for y_intercept.**

- **First, we factored y_hat to (-mxi-b) in the error function `sum of squared errors`**
- **Then we took the partial derivative of the whole equation with respect to `b`**
- **applying chain rule, d/db(-b) = -1 and d/dsse (f(sse))^2 = 2(f(see))^(2-1)**
- **multiplying both, -2(f(sse)) then we set it equal to zero**
- **then we divided both sides by -2**

- **we can distribute sigma in the equation and sigma of b is just n*b**
- **next, we find b by movind yi and mxi to the other side then divide both sides by n**
- **if we divide sigma(yi) by n and the same for sigma(xi), we get the mean for both**
- **we obtain an equation that if used we can easly use to obtain y-intercept b**

###  Slope of the line `m`

**Applying the same steps for fiding b, we obtained equation for calculating slope of a line that fits the data `m`.**


### C- Implementing Cost Function 

**Now that we are done with the hard part, it is time to code the equation we derived for slope and y-intercept 
for our example of `engine-size` and `price`.**



```python
# define splitting function 
# test train split
def split_array(x, y):
    shuffle = np.random.rand(x.shape[0])
    partition = shuffle < np.percentile(shuffle, 80)

    x_train = x[partition]
    y_train = y[partition]
    x_test = x[~partition]
    y_test = y[~partition]

    return x_train, y_train, x_test, y_test

# define function to calculate slope
def slope(x,y):
    numerator = np.sum(x*y - np.mean(y)*x)
    denomirator = np.sum(np.power(x,2) - np.mean(x)*x)
    return numerator / denomirator

def y_intercept(x,y):
    return np.mean(y) - slope(x,y)*np.mean(x)

# define predicting function 
def y_pred(x,m,b):
    return m*x + b

# define error function 
def error(y,y_hat):
    return np.sum(np.power(y-y_hat,2))
```

```python
# obtain training and testing data
x_train,y_train,x_test,y_test = split_array(df['engine-size'].values,df['price'].values)
```

```python
# calculating slope 'm' 
m = slope(x_train,y_train)
print('the slope of the line that fits engine-size and price:')
print('\u2022',m)
```

```python
# calculating y_intercept 'b'
b = y_intercept(x_train,y_train)
print('the y_intercept of the line that fits engine-size and price:')
print('\u2022',b)
```

### Now we have an equation of a line `y_hat` that we can use to estimate new values

**y_hat = 152.447873*x -6355.811195**
- y_hat = predicted car price
- x = engine-size

Let us calculate y_hat using test data we got from `split_array` function as if they are new data for other cars. 

****Note: Rerunning the functions we result in different `m` and `b` results since `random shafful` is not repreducable. Use np.random.seed method to lay reproducable slices of x_test/train and y_test/train data**

```python
# calculating y_hat on x_test data
y_hat = y_pred(x_test,m,b)
y_hat[:5]
```

```python
# calculating the error 'residual'
sse = error(y_test,y_hat)
print('sum of squared errors for testing data:',sse)
```
